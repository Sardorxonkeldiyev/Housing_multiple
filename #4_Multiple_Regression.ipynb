```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/housing_LR.csv', index_col=0)
df.head(10)
```

```python
housing = df[df.district == 'Чиланзарский'] # bu kodimiz df dagi district ustunidagi 'Чиланзарский' shu malumotlar qatorlari va ustunlarini ajratadi
housing.shape
```

# Train_test split


```python
from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(housing,test_size=0.10, random_state=42)
```

# ML (Machine learning) :

```python
from sklearn import linear_model
MLR_model = linear_model.LinearRegression()
# x_train:
x_train = np.asanyarray(train_set[['rooms','size','level','max_levels']])

# y_train :
y_train = np.asanyarray(train_set[['price']])

# MLR modelni train qilamiz:
MLR_model.fit (x_train, y_train)

# Biz aynan qaysi ustun 'price' ustuniga yaqin ekanligini ham sonlar orqali ko'tishimiz mumkin, bunda qaysi ustunni raqamlari katta bo'lsa usha yaqin-->
#--> hisoblanadi :

print('Coeffsientlar:', MLR_model.coef_)
print('theta0:', MLR_model.intercept_)
```

 * yuqorida ko'rib turganimizdek ikkinchi ustunni raqamlari ktta yani 862.93 chiqdi bu 'size' ustuniga to'gri keladi va 148.01 'max_levels' ustunini ham modelga qushsak boladi

```python
# Endi esa 'price' ustunlaribn o'zaro bog'liqlik da bo'lgan ustunni grafik orqali ko'z bn aniqlaymiz :

sns.pairplot(df[['rooms','size','level','max_levels','price']])
plt.show()
```

# Modelni baholaymiz :

```python
x_test = np.asanyarray(test_set[['rooms','size','level','max_levels']])
y_test = np.asanyarray(test_set[['price']])
y_predict = MLR_model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

MAE = mean_absolute_error(y_test, y_predict)
RMSE = np.sqrt(mean_squared_error(y_test, y_predict))
print(f"{MAE=}")
print(f"{RMSE=}")
```
